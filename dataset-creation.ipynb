```python
import pandas as pd

df = pd.read_csv("data/annotations/annotations.tsv", sep='\t', header=None)
df.head()
```

```python
df["items"] = df[df.columns[1:]].values.tolist()
df["path"] = df[df.columns[0]]
df["scene"] = df.apply(lambda x: x["path"].split("/")[0], axis=1)
df["path"] = df.apply(lambda x: "./data/imageAlignedLD/" + x["path"], axis=1)
df.head()
```

```python
df2 = df[["path", "scene", "items"]]
```

```python
df2.head().path.tolist()
```

```python
import numpy as np

attributes = [("night", 2), ("snow", 9), ("rain", 29), ("sunny", 6), ("cloudy", 7)]

def map_values(target_idx, threshold=0.9):
    def _mapper(row):
        items = row["items"]
        vals = [float(i.split(",")[0]) for i in items]
        return vals[target_idx] >= threshold
    return _mapper

threshold=0.95
df2["is_night"] = df2.apply(map_values(2), axis=1)
df2["is_snow"] = df2.apply(map_values(9), axis=1)
df2["is_rain"] = df2.apply(map_values(29), axis=1)
df2["is_sunny"] = df2.apply(map_values(6), axis=1)
df2["is_cloudy"] = df2.apply(map_values(7), axis=1)
df3 = df2[(df2.is_night | df2.is_snow | df2.is_rain | df2.is_sunny | df2.is_cloudy)]
```

```python
labels = ['cloudy', 'night', 'rain', 'snow', 'sunny']
stoi = {s: i for i, s in enumerate(labels)}
itos = {i: s for i, s in enumerate(labels)}

def map_labels(row):
    if row.is_cloudy:
        return 0
    elif row.is_night:
        return 1
    elif row.is_rain:
        return 2
    elif row.is_snow:
        return 3
    elif row.is_sunny:
        return 4

df3["label"] = df3.apply(map_labels, axis=1)
stoi, itos
```

```python
from datasets import Dataset

main_df = df3[["path", "label"]]
main_ds = Dataset.from_pandas(main_df).select_columns(["path", "label"])
main_ds = main_ds.class_encode_column("label")
```

```python
main_ds = main_ds.train_test_split(test_size=0.2, stratify_by_column="label")
```

```python
from PIL import Image
import random
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 3, figsize=(25, 25))

flattened_axes = [
    x
    for xs in axs
    for x in xs
]
        
for ax in flattened_axes:
    item = random.choice(main_ds["train"])
    img = Image.open(item["path"])
    ax.imshow(img)
    ax.set_title(itos[item["label"]])
plt.tight_layout()
```
